# Exploratory Data Analysis

### Import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

### Load dataset into memory

```python
dataset = load_breast_cancer()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
df.head()
```

### Check for presence of any NA values

```python
df.isna().sum()
```

### Separate features and target

```python
X = df.drop('target', axis='columns')
y = df.target
```

```python
X.shape
```

```python
y.shape
```

### Check datatypes of each feature

```python
X.info()
```

### Perform basic statistical analysis

```python
X.describe()
```

### Check distribution of classes in target

```python
y.map({0: 'M', 1: 'B'}).value_counts().plot(kind='bar')
```

```python
y.value_counts()
```

### Analyse distribution of every feature separetely for each class

```python
for column in X:
    plt.figure()
    column_data = X[column]
    sns.kdeplot(column_data[y == 0], label='B')
    sns.kdeplot(column_data[y == 1], label='M')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title('Distribution of {}'.format(column))
```

### Calculate and visualise correlations between features

```python
plt.figure(figsize=(20, 16))
sns.heatmap(X.corr(), annot=True, cmap='Spectral')
```
